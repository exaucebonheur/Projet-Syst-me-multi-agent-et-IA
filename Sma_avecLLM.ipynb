```python
# -*- coding: utf-8 -*-
"""Sma_avecLLM.ipynb
Version corrig√©e et s√©curis√©e
"""

# Installation des d√©pendances
!pip install mesa==2.1.0
!pip install transformers torch matplotlib huggingface_hub python-dotenv

# Imports
from mesa import Agent, Model, time
from mesa.space import MultiGrid
from mesa.datacollection import DataCollector
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import random
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv
from IPython.display import display, HTML
import textwrap

# Ajuster la largeur de la sortie texte
display(HTML("<style>pre { white-space: pre-wrap; }</style>"))

# ‚úÖ Charger le token Hugging Face depuis .env
from google.colab import files
uploaded = files.upload()  # uploade ton fichier .env depuis ton PC

import os
from dotenv import load_dotenv

load_dotenv(".env")  # charge ton token
token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
print("‚úÖ Token charg√© :", token[:10] + "...")

# Charger le mod√®le Mistral
model_name = "mistralai/Mistral-7B-Instruct-v0.1"

# Charger le tokenizer et le mod√®le avec float16
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    use_auth_token=token,
    torch_dtype=torch.float16,  # Forcer FP16 pour r√©duire la m√©moire
    device_map="auto"  # GPU si dispo
)

# V√©rifier le chargement du mod√®le
print("‚úÖ Mod√®le charg√© sur :", model.hf_device_map)


# Fonction d‚Äôanalyse du sentiment
def analyze_sentiment(text):
    prompt = f"Analyse la situation suivante du march√© et donne une recommandation : \n{text}"

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(**inputs, max_length=100, num_return_sequences=1)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    sentiment = "hausse" if "hausse" in response else "baisse"

    return sentiment, textwrap.fill(response, width=80)


# Classe Agent
class SmartTrader(Agent):
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.wealth = 100

    def step(self):
        market_news = [
            "Le march√© est en hausse aujourd'hui.",
            "Le march√© est en baisse aujourd'hui.",
            "Le march√© est stable aujourd'hui.",
            "Le march√© est volatile aujourd'hui.",
            "Le march√© est incertain aujourd'hui."
        ]
        news = random.choice(market_news)

        sentiment, response = analyze_sentiment(news)

        print(f"\nüìä Agent {self.unique_id} analyse :\n{response}\n")

        if sentiment == "hausse" and self.wealth > 20:
            if random.random() < 0.7:
                print(f"üü¢ Agent {self.unique_id} d√©cide d'acheter.")
                self.wealth -= 10
            else:
                print(f"‚ö™ Agent {self.unique_id} ne fait rien.")
        elif sentiment == "baisse" and self.wealth < 150:
            if random.random() < 0.7:
                print(f"üî¥ Agent {self.unique_id} d√©cide de vendre.")
                self.wealth += 10
            else:
                print(f"‚ö™ Agent {self.unique_id} ne fait rien.")
        else:
            print(f"‚ö™ Agent {self.unique_id} ne fait rien.")


# Classe Mod√®le
class SmartMarketModel(Model):
    def __init__(self, N):
        self.num_agents = N
        self.grid = MultiGrid(10, 10, True)
        self.schedule = time.RandomActivation(self)
        for i in range(self.num_agents):
            trader = SmartTrader(i, self)
            self.schedule.add(trader)
            x = random.randrange(self.grid.width)
            y = random.randrange(self.grid.height)
            self.grid.place_agent(trader, (x, y))

    def step(self):
        self.schedule.step()


# Simulation
smart_model = SmartMarketModel(5)
for i in range(2):
    print(f"\n=== Step {i+1} ===")
    smart_model.step()

# R√©cup√©rer les valeurs de richesse
smart_wealth_values = [agent.wealth for agent in smart_model.schedule.agents]

# Histogramme
plt.hist(smart_wealth_values, bins=20, range=(60, 140), density=True)
plt.title("Distribution de la richesse des traders (avec Mistral)")
plt.xlabel("Richesse")
plt.ylabel("Proportion de traders")
plt.show()
```
